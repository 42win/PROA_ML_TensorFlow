##Jenis2 Format

formating +

- ("words " + variabel)

```python
print("Halo, namaku " + namaku)
print("Umurku " + umurku)
```

- formating %

  - kekurangannya sulit dibaca dan jika ingin menggunakan variabel yg sama kita tetap harus menuliskannya lagi ex: umurku, umurku..

```python
# %-Formatting
print("Halo, namaku %s.\nUmurku %s. %s umur %s" % (namaku, umurku, namaku, umurku))
```

- formating str

  - hampir sama dengan formating sebelumnya namun kita tidak perlu menulis ulang variabel jika ingin dipakai kembali
  - serta sudah lebih mudah dibaca variabel yg digunakan

  - tapi ini tetap menjadi kekurangan jika variabelnya sudah terlalu banyak

```python
# str.format()
# 0 itu pertama, 1 itu kedua
print("Halo, namaku {nama}.\nUmurku {umur}.\n{nama} umur {umur}".format(nama = namaku, umur = umurku))
```

- formating f-string

  - default format yg sering dipakai
  - langsung mengambil dari variabelnya
  - memperbaiki kekurangan formating sebelumnya 

  - bisa langsung melakukan operasi didalamnya

```python
# f-String
print(f"Halo, namaku {namaku}.\nUmurku {umurku}.\n{namaku} umur {umurku}")
print(f"Halo, namaku {namaku}.\nUmurku {umurku}.\n{namaku} umur {tahunSekarang - tahunKelahiran}")
print(f"{{{tahunSekarang - tahunKelahiran}}}")
```

## Perbedaan Tuple, Set, List

```python
# Tuple - Gak bisa berubah, gak bisa nambah, ada order
a = (1,2,3)

# a= 1,2,3 bisa juga tidak pakai kurung

print(a)
print(type(a))

# Set - Gak bisa diubah, bisa nambah, gak ada order (urutan), no duplikat
b = {1,2,3}
print(b)
print(type(b))

# List - Bisa ubah, bisa nambah, order, duplikat
c = [1,2,3]
print(c)
print(type(c))
```

contoh penggunaan 
- tuple, misal data kordinat
- set, nomor ktp


- Dictionary data type

  - hampir sama dengan set tapi dictionary bisa diubah valuenya
  - key value concept 

```python
# Dictionary - Bisa diubah, bisa nambah, gak ada order, no duplikat
#          Key : Value
kamus = {"nama": "Jamal",
         "umur":19,
         "laki":True,
         "teman": ["Gugun", "Janico", "Abdul"]
         }
kamus["nama"] = "Zain"
print(kamus["teman"][0])
print(type(kamus))
```
